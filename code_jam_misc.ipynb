## Alien Language

### Problem Description
After years of study, scientists at Google Labs have discovered an alien
language transmitted from a faraway planet. The alien language is very
unique in that every word consists of exactly L lowercase letters. Also,
there are exactly D words in this language.

Once the dictionary of all the words in the alien language was built,
the next breakthrough was to discover that the aliens have been
transmitting messages to Earth for the past decade. Unfortunately, these
signals are weakened due to the distance between our two planets and some
of the words may be misinterpreted. In order to help them decipher these
messages, the scientists have asked you to devise an algorithm that will
determine the number of possible interpretations for a given pattern.

A pattern consists of exactly L tokens. Each token is either a single
lowercase letter (the scientists are very sure that this is the letter)
or a group of unique lowercase letters surrounded by parenthesis ( and ).
For example: (ab)d(dc) means the first letter is either a or b, the second
letter is definitely d and the last letter is either d or c. Therefore,
the pattern (ab)d(dc) can stand for either one of these 4 possibilities:
add, adc, bdd, bdc.

Input

The first line of input contains 3 integers, L, D and N separated by a
space. D lines follow, each containing one word of length L. These are
the words that are known to exist in the alien language. N test cases
then follow, each on its own line and each consisting of a pattern as
described above. You may assume that all known words provided are unique.

Output

For each test case, output

Case #X: K
where X is the test case number, starting from 1, and K indicates how
many words in the alien language match the pattern.

Limits

Small dataset

1 ≤ L ≤ 10
1 ≤ D ≤ 25
1 ≤ N ≤ 10
Large dataset

1 ≤ L ≤ 15
1 ≤ D ≤ 5000
1 ≤ N ≤ 500

## All Your Base

### Problem Description

Problem

In A.D. 2100, aliens came to Earth. They wrote a message in a cryptic 
language, and next to it they wrote a series of symbols. We've come to 
the conclusion that the symbols indicate a number: the number of seconds 
before war begins!

Unfortunately we have no idea what each symbol means. We've decided that 
each symbol indicates one digit, but we aren't sure what each digit means 
or what base the aliens are using. For example, if they wrote "ab2ac999", 
they could have meant "31536000" in base 10 -- exactly one year -- or they 
could have meant "12314555" in base 6 -- 398951 seconds, or about four and 
a half days. We are sure of three things: the number is positive; like us, 
the aliens will never start a number with a zero; and they aren't using 
unary (base 1).

Your job is to determine the minimum possible number of seconds before war 
begins.

Input

The first line of input contains a single integer, T. T test cases follow. 
Each test case is a string on a line by itself. The line will contain only 
characters in the 'a' to 'z' and '0' to '9' ranges (with no spaces and no 
punctuation), representing the message the aliens left us. The test cases 
are independent, and can be in different bases with the symbols meaning 
different things.

Output

For each test case, output a line in the following format:

Case #X: V
Where X is the case number (starting from 1) and V is the minimum number 
of seconds before war begins.
Limits

1 ≤ T ≤ 100
The answer will never exceed 1018
Small dataset

1 ≤ the length of each line < 10

Large dataset

1 ≤ the length of each line < 61

Sample

Input           Output 
3
11001001        Case #1: 201
cats            Case #2: 75
zig             Case #3: 11

### Solution

```python
def decrypt(code):
    """
    INPUT - a list of single characters such as ['1','a','2','3','c']
    OUTPUT - modifies the list itself to such as [1,0,2,3,1]
    'Decryptes' the code by interpreting each character as whatever digit
    leads to the smallest number
    """
    values = {}
    possNums = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]
    counter = 0
    for char in code:
        try:
            code[counter] = int(char)
                
        except ValueError: #when it's a letter
        #the first char can't be 0
            if char == code[0]:
                code[0] = 1
                values[code[0]] = 1
                possNums.remove(1)
                
            elif char not in values:
                code[counter] = possNums.pop()
                values[char] = code[counter]
                                        
            else:
                code[counter] = values[char]
                
        counter += 1
        

def FBAdd(args):
    """
    (FindBase&Add)
    INPUT - A list of integers
    OUTPUT - A single integer
    Returns the number resulting from interpreting the list as a 
    single integer with whichever base makes it the smallest
    Does not modify the original list
    """
    ints = list(args)
    base = max(max(ints) + 1, 2) #cannot be unary
    maxPower = len(ints)  #start with 0th power
    answer = 0
    for power in range(maxPower):
        answer += ints.pop()*base**power
    return answer
    
        

def run(verbose = False):
    size = raw_input("Enter S for small or L for large: ")
    if size == "S":
        inFile = open("All Your Base Small Input.in")
        outFile = open("All Your Base Small Output.txt", 'w')
        
    elif size == "L":
        inFile = open("All Your Base Large Input.in")
        outFile = open("All Your Base Large Output.txt", 'w')
    
    cases = int(inFile.readline())
    for case in range(1, cases+1):
    #Convert to a list to more easily interact with and change
    #individual elements
        code = list(inFile.readline().strip('\n'))
        if verbose: print "Code before decryption: %s" % (code)
        decrypt(code)
        answer = FBAdd(code)
    
        if verbose: print "Case #%s: %s" % (case, answer)
        if verbose: print "Code after decryption: %s \n" % (code)
    
        outFile.write("Case #%s: %s\n" % (case, answer))

    inFile.close()
    outFile.close()
    
#I seem to be missing something...small input turned out wrong
#Large input had more than ten different letters, which means that
#each number 0-9 isn't just used once...I don't understand the specs then
```

## Deceitful War

### Problem Description
Naomi and Ken sometimes play games together. Before they play, each of them 
gets N identical-looking blocks of wood with masses between 0.0kg and 1.0kg 
(exclusive). All of the blocks have different weights. There are lots of games 
they could play with those blocks, but they usually play something they call 
War. Here is how War works:

Each player weighs each of his or her own blocks, so each player knows the 
weights of all of his or her own blocks, but not the weights of the other 
player's blocks.

They repeat the following process N times:
Naomi chooses one of her own blocks, with mass ChosenNaomi.
Naomi tells Ken the mass of the block she chose.
Ken chooses one of his own blocks, with mass ChosenKen.
They each put their block on one side of a balance scale, and the person whose 
block is heavier gets one point.
Both blocks are destroyed in a fire.
Naomi has realized three things about War. First, she has realized that she 
loses a lot. Second, she has realized that there is a unique strategy that Ken 
can follow to maximize his points without assuming anything about Naomi's 
strategy, and that Ken always uses it. Third, she has realized that she hates 
to lose. Naomi has decided that instead of playing War, she will play a game 
she calls Deceitful War. The great thing about Deceitful War is that Ken will 
think they're playing War!

Here is how Deceitful War works, with differences between Deceitful War and 
War in bold:

Each player weighs each of his or her own blocks. Naomi also weighs Ken's 
blocks while he isn't looking, so Naomi knows the weights of all blocks and 
Ken only knows the weights of his own blocks.

They repeat the following process N times:
Naomi chooses one of her own blocks, with mass ChosenNaomi.
Naomi tells Ken a number, ToldNaomi, between 0.0kg and 1.0kg exclusive. Ken, 
who thinks they're playing War, thinks the number Naomi just told him is 
ChosenNaomi.
Ken chooses one of his own blocks, with mass ChosenKen.
They each put their block on one side of a balance scale, and the person whose 
block is heavier gets one point.
Both blocks are destroyed in a fire.
Naomi doesn't want Ken to know that she isn't playing War; so when she is 
choosing which block to play, and what mass to tell Ken, she must make sure 
that the balance scale won't reveal that ChosenNaomi ≠ ToldNaomi. In other 
words, she must make decisions so that:

ChosenNaomi > ChosenKen if, and only if, ToldNaomi > ChosenKen, and
ToldNaomi is not equal to the mass of any of Ken's blocks, because he knows 
that isn't possible.
It might seem like Naomi won't win any extra points by being deceitful, because
Ken might discover that she wasn't playing War; but Naomi knows Ken thinks both
players are playing War, and she knows what he knows, and she knows Ken will 
always follow his unique optimal strategy for War, so she can always predict 
what he will play.

You'll be given the masses of the blocks Naomi and Ken started with. Naomi 
will play Deceitful War optimally to gain the maximum number of points. Ken 
will play War optimally to gain the maximum number of points assuming that both
players are playing War. What will Naomi's score be? What would it have been if
she had played War optimally instead?

Examples

If each player has a single block left, where Naomi has 0.5kg and Ken has 0.6kg,
then Ken is guaranteed to score the point. Naomi can't say her number is ≥ 
0.6kg, or Ken will know she isn't playing War when the balance scale shows his 
block was heavier.

If each player has two blocks left, where Naomi has [0.7kg, 0.2kg] and Ken has 
[0.8kg, 0.3kg], then Naomi could choose her 0.2kg block, and deceive Ken by 
telling him that she chose a block that was 0.6kg. Ken assumes Naomi is telling
the truth (as in how the War game works) and will play his 0.8kg block to score
a point. Ken was just deceived, but he will never realize it because the balance
scale shows that his 0.8kg block is, like he expected, heavier than the block 
Naomi played. Now Naomi can play her 0.7kg block, tell Ken it is 0.7kg, and 
score a point. If Naomi had played War instead of Deceitful War, then Ken would
have scored two points and Naomi would have scored zero.

Input

The first line of the input gives the number of test cases, T. T test cases 
follow. Each test case starts with a line containing a single integer N, the 
number of blocks each player has. Next follows a line containing N 
space-separated real numbers: the masses of Naomi's blocks, in kg. Finally 
there will be a line containing N space-separated real numbers: the masses of 
Ken's blocks, in kg.

Each of the masses given to Ken and Naomi will be represented as a 0, followed 
by a decimal point, followed by 1-5 digits. Even though all the numbers in the 
input have 1-5 digits after the decimal point, Ken and Naomi don't know that; 
so Naomi can still tell Ken that she played a block with mass 0.5000001kg, and 
Ken has no reason not to believe her.

Output

For each test case, output one line containing "Case #x: y z", where x is the 
test case number (starting from 1), y is the number of points Naomi will score 
if she plays Deceitful War optimally, and z is the number of points Naomi will 
score if she plays War optimally.

Limits

1 ≤ T ≤ 50.
All the masses given to Ken and Naomi are distinct, and between 0.0 and 1.0 
exclusive.
Small dataset

1 ≤ N ≤ 10.
Large dataset

1 ≤ N ≤ 1000.

```python
def create_grid(grid_file):
    """
    :param grid_file: A text file containing the grid; the following lines
    (those describing the next grid) will be read off
    :return: a grid as a list of lists (list of grid rows) along with the size
    of the grid and the size (length) of a series of R/B pieces needed in the
    grid for a victory
    """
    grid_size, victory_size = (int(x) for x in grid_file.readline().split())

    grid = []
    for k in range(grid_size): grid.append([])

    for row in range(grid_size):
        grid[row].append(list(inFile.readline().strip()))
    return grid, grid_size, victory_size


def rotate(grid):
    """
    :param grid: a list of lists. Each list represents one row in the grid.
    Original remains untouched.
    The grid is rotated 90 degrees clockwise.
    :return: a new list containing the rotated grid;
    """

    rotated_grid = []
    for k in range(len(grid)): rotated_grid.append([])

    for i in range(len(grid)):
        #iterate backwards through the rows (for 90 degree rotation)
        for j in range(len(grid)-1, -1, -1):
            rotated_grid[i].append(grid[j])
    return rotated_grid


def freefall(grid):
    """
    :param grid: a list of lists. Each list represents one row in the grid.
    Original remains untouched.
    :return: a new list in which all of the pieces have fallen down according
    to gravity
    """
    fallen_grid = grid[:]

#are you accounting for the lowest empty position possibly being different in each column?
    for column in range(len(fallen_grid)):
        #the lowest position is in the last list (last row)
        lowest = len(fallen_grid) - 1
        for row in range(len(fallen_grid)-1, -1, -1):
            #start at the bottom and move up; if there's ever a non-blank
            #square, move it down to lowest blank space
            if fallen_grid[row][column] == 'R' or 'B':
                #check below; remember, higher index is lower in grid
                if lowest > column:
                    fallen_grid[row][lowest] = fallen_grid[row][column]
                    fallen_grid[row][column] = '.'
                lowest -= 1  #lowest space already full or else just barely
                #filled; next lowest is above
    return fallen_grid


def diagonal_from(index, grid):
    """
    :param index: A tuple containing the (row, column) index of a node
    :param grid:
    :return: A list of all nodes in grid diagonal from node at index
    """
    for i in range(-len(grid), len(grid)):
        for j in range(-len(grid), len(grid)):
            if abs(i) == abs(j):
                pass


def calculate_win(grid, victory_size):
    """
    The win condition is to have victorySize pieces of the same color in a row,
    column, or diagonal
    :param grid:
    :return:
    """
    max_blue_length = 0
    max_red_length = 0

    #check all rows
    for row in range(len(grid)):
        current_red_length = 0
        current_blue_length = 0
        for elem in grid[row]:
            if elem == 'R':
                current_red_length += 1
                current_blue_length = 0
            elif elem == 'B':
                current_blue_length += 1
                current_red_length = 0
            max_blue_length = max(max_blue_length, current_blue_length)
            max_red_length = max(max_red_length, current_red_length)

        #check in columns
        for column in range(len(grid)):
            current_red_length = 0
            current_blue_length = 0
            for row in range(len(grid)):
                if grid[row][column] == 'R':
                    current_red_length += 1
                    current_blue_length = 0
                elif grid[column][row] == 'B':
                    current_blue_length += 1
                    current_red_length = 0
                max_blue_length = max(max_blue_length, current_blue_length)
                max_red_length = max(max_red_length, current_red_length)

        #check diagonals
        diagonal_rows = []
        #make them from the first row down and last row up
        index = 0
        for elem in grid[0]:
            diagonal_rows.append([elem] + diagonal_from([0, index]))
            index += 1

        for row in diagonal_rows:
            current_red_length = 0
            current_blue_length = 0
            for elem in row:
                if elem == 'R':
                    current_red_length += 1
                    current_blue_length = 0
                elif elem == 'B':
                    current_blue_length += 1
                    current_red_length = 0
                max_blue_length = max(max_blue_length, current_blue_length)
                max_red_length = max(max_red_length, current_red_length)

    if max_blue_length >= victory_size and max_red_length >= victory_size:
        return 'Both'
    elif max_blue_length >= victory_size:
        return 'Blue'
    elif max_red_length >= victory_size:
        return 'Red'
    else: return 'Neither'


input = 'small_input.in'
#input = 'large_input.in'

outFile = open('small_output.txt', 'w')
#outFile = open('large_output.txt', r)

inFile = open(input)
numTestCases = int(inFile.readline())

for case in range(1, numTestCases+1):
    grid, gridSize, victorySize = create_grid(inFile)

    outcome = calculate_win(freefall(rotate(grid)), victorySize)

    outFile.write('Case #%s: %s' % (numTestCases, outcome))


inFile.close()
outFile.close()
```

## Watersheds

### Problem Description
Geologists sometimes divide an area of land into different regions based on where rainfall flows down to. These regions are called drainage basins.

Given an elevation map (a 2-dimensional array of altitudes), label the map such that locations in the same drainage basin have the same label, subject to the following rules.

From each cell, water flows down to at most one of its 4 neighboring cells.
For each cell, if none of its 4 neighboring cells has a lower altitude than the current cell's, then the water does not flow, and the current cell is called a sink.
Otherwise, water flows from the current cell to the neighbor with the lowest altitude.
In case of a tie, water will choose the first direction with the lowest altitude from this list: North, West, East, South.
Every cell that drains directly or indirectly to the same sink is part of the same drainage basin. Each basin is labeled by a unique lower-case letter, in such a way that, when the rows of the map are concatenated from top to bottom, the resulting string is lexicographically smallest.
```python
class NoSinkError(Exception):
    ''''''

class Tile(object):
    def __init__(self, x, y, altitude):
        assert type(x) == type(y) == type(altitude) == int
        self.x = x
        self.y = y
        self.alt = altitude
        self.isSink = False
        self.sink = None
        self.letter = None

    def getNeighbors(self, Map):
        x = self.x
        y = self.y
        neighbors = [Map.get((x,y+1)), Map.get((x,y-1)), Map.get((x-1,y)), Map.get((x+1,y))]
        toDel = []
        for i in range(len(neighbors)):
            if type(neighbors[i]) == type(None):
                toDel.append(neighbors[i])
        for elem in toDel:
            neighbors.remove(elem)
        return neighbors

    def findSink(self, Map):
        neighbors = self.getNeighbors(Map)
        sinkNeighbors = []
        
        for neighbor in neighbors:
            if neighbor.isSink:
                sinkNeighbors.append(neighbor)
        
        for neighbor in neighbors:
            if sinkNeighbors:
                self.sink = neighbor
                return neighbor
            
            else:
                minNeighbor = None
                minAlt = self.alt
                
                for neighbor in neighbors:
                    if neighbor.alt < minAlt:
                        minAlt = neighbor.alt
                        minNeighbor = neighbor
                        ties = [neighbor]
                        
                    elif neighbor.alt == minAlt and minAlt != self.alt:
                        ties.append(neighbor)
                        
                if minAlt != self.alt:
                    self.sink = minNeighbor.findSink(Map)
                    return self.sink
                
                else:
                    #this tile is a sink
                    self.isSink = True
                    #return self
                    return self

    def setLetter(self):
        if self.sink != None:
            self.letter = self.sink.letter
        else:
            raise NoSinkError

    def __repr__(self):
        if self.isSink:
            return 'SINK; x = %s, y = %s, alt = %s, sink = %s, letter = %s' % (self.x, self.y, self.alt, self.sink, self.letter)
        else:
            return 'Tile; x = %s, y = %s, alt = %s, sink = %s, letter = %s' % (self.x, self.y, self.alt, self.sink, self.letter)
            
    def __eq__(self, other):
        if type(other) == type(None):
            return False
        
def cmpPos(tile1, tile2):
    '''To sort the sinks; starts at top left, moves down columns then right'''
    if tile1.x < tile2.x:
        return -1
    elif tile2.x < tile1.x:
        return 1
    elif tile1.x == tile2.x:
        if tile1.y < tile2.y:
            return -1
        elif tile2.y < tile1.y:
            return 1
        
def mapDrainages(filename):
    inFile = open(filename)
    res =''
    
    numMaps = int(inFile.readline())
    for case in range(numMaps):
        height, width = [int(elem) for elem in inFile.readline().split(' ')]

        #dictionary of positions (x,y) --> Tiles
        Map = {}
        for yPos in range(height):
            row = [int(elem) for elem in inFile.readline().split(' ')]
            for xPos in range(len(row)):
                Map[xPos, yPos] = Tile(xPos, yPos, row[xPos])

        #alters the Map to replace Tiles with Basins where appropriate
        positions = Map.keys()
        for pos in positions:
            tile = Map[pos]
            if tile.sink == None:
                tile.findSink(Map)
            
        sinks = sorted([tile for tile in Map.values() if tile.isSink == True], cmpPos)
        alphabet = 'abcdefghijklmnopqrstuvwxyz'
        for i in range(len(sinks)):
            sinks[i].letter = alphabet[i]
        for tile in Map.values():
            if not tile.isSink:
                tile.setLetter()
                
        res += 'Case #%s:\n' % (case + 1)
        for j in range(height):
            for i in range(width):
                res += Map[i,j].letter + ' '
            res += '\n'
        print case
    return res


filename = 'Watersheds Small Input.in'
outFile = open('Watersheds Small Output.txt', 'w')
outFile.write(mapDrainages(filename))
outFile.close()

##TODO: update so that minNeighbor gives the proper neighbor if two are equal;
##look at spec to see how to handle this

##file2 = 'Watersheds Large Input.in'
##outFile = open('Watersheds Large Output.txt', 'w')
##outFile.write(mapDrainages(file2))
##outFile.close()



##    def __lt__(self, other):
##        assert isinstance(other, Tile)
##        print 'used'
##        return self.alt < other.alt
##    def __gt__(self, other):
##        assert isinstance(other, Tile)
##        return self.alt > other.alt
##    def __le__(self, other):
##        assert isinstance(other, Tile)
##        return self.alt < other.alt or self.alt == other.alt
##    def __ge__(self, other):
##        assert isinstance(other, Tile)
##        return self.alt > other.alt or self.alt == other.alt
##    def __ne__(self, other):
##        if type(other) == type(None):
##            return True
##        assert isinstance(other, Tile)
##        return self.alt != other.alt
```
